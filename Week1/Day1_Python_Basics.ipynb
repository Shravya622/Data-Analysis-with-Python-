## Mini Project2 ‚Äì Grade Calculator


DAY 1 MINI PROJECT

Build this:

üìå Simple Grade Calculator

Program should:

1Ô∏è‚É£ Take student name
2Ô∏è‚É£ Take 3 subject marks
3Ô∏è‚É£ Calculate average
4Ô∏è‚É£ Print:

Shravya scored average 85.6
Grade: A

Grading logic:

90+ ‚Üí A+

80‚Äì89 ‚Üí A

70‚Äì79 ‚Üí B

60‚Äì69 ‚Üí C

Below 60 ‚Üí Fail

```python
student_name=input("Enter the Student Name:")
sub_1=int(input("Enter the Marks Obtained in Maths:"))
sub_2=int(input("Enter the Marks Obtained in Physics:"))
sub_3=int(input("Enterthe Marks Obtained in Chemistry:"))

average=(sub_1+sub_2+sub_3)/3
print(f"{student_name} scored average {average:.2f}")
if average>=90:
  print("Grade:A+")
elif average>=80:
  print("Grade:A")
elif average>=70:
  print("Grade:B")
elif average>=60:
  print("Grade:C")
else:
  print("Fail")
```

##Mini Project3 ‚Äì Scholarship eligibility Checker

Build this:

üìå Smart Eligibility Checker

Program should:

1Ô∏è‚É£ Take name
2Ô∏è‚É£ Take age
3Ô∏è‚É£ Take percentage

Conditions:

If age ‚â• 18 AND percentage ‚â• 60
‚Üí "Eligible for Scholarship"

If age ‚â• 18 AND percentage < 60
‚Üí "Not eligible due to low percentage"

If age < 18
‚Üí "Not eligible due to age"

Use:

input

int

float

and

if-elif-else

f-strings

```python
name = input("Enter your Name: ")
age = int(input("Enter your age: "))
percentage = float(input("Enter your percentage: "))

if age >= 18 and percentage >= 60:
    print(f"{name}, Eligible for Scholarship")

elif age >= 18 and percentage < 60:
    print(f"{name}, Not eligible due to low percentage")

else:
    print(f"{name}, Not eligible due to age")
```
