```python
numbers = [12, 45, 67, 89, 34, 22]

total = 0

for num in numbers:
    total += num

average = total / len(numbers)

print(f"Total = {total}")
print(f"Average = {average:.2f}")

print("\nNumbers greater than 50:")

for num in numbers:
    if num > 50:
        print(num)
```

**Processing Dictionary Data**

```python
marks={
    "physics":90,
    "chemistry":95,
    "maths":96,
    "Biology":99
}

for subject,marks in marks.items():
  print(f"{subject}:{marks}")
```

**List of Dictionaries (Dataset Structure)**

```python
students = [
    {"name": "Shravya", "marks": 90},
    {"name": "Anu", "marks": 75},
    {"name": "Rahul", "marks": 85},
    {"name": "Riya", "marks": 60}
]
```

**Practice 3**

Using the dataset above:

Write code to:

    Print all student names

    Find the student with highest marks

    Print students with marks ≥ 80

```python
students = [
    {"name": "Shravya", "marks": 90},
    {"name": "Anu", "marks": 75},
    {"name": "Rahul", "marks": 85},
    {"name": "Riya", "marks": 60}
]

print("Student Names:")
for student in students:
    print(student["name"])

top = students[0]
for student in students:
    if student["marks"] > top["marks"]:
        top = student

print("\nTop Scorer:")
print(f"{top['name']} - {top['marks']}")

print("\nStudents with marks ≥ 80:")
for s in students:
    if s["marks"] >= 80:
        print(f"{s['name']} - {s['marks']}")


```

##Mini Project — Student Dataset Analyzer

**Dataset:**

students = [
    {"name": "Shravya", "marks": 90},
    {"name": "Anu", "marks": 75},
    {"name": "Rahul", "marks": 85},
    {"name": "Riya", "marks": 60}
]

 Tasks:

    Calculate average marks

    Print students who passed (marks ≥ 70)

    Print top scorer

```python
students = [
    {"name": "Shravya", "marks": 90},
    {"name": "Anu", "marks": 75},
    {"name": "Rahul", "marks": 85},
    {"name": "Riya", "marks": 60}
]

total = 0
for s in students:
    total += s["marks"]

average = total / len(students)
print("Average marks:", average)

print("\nStudents who passed:")
for s in students:
    if s["marks"] >= 70:
        print(s["name"], "-", s["marks"])

top = students[0]
for s in students:
    if s["marks"] > top["marks"]:
        top = s

print("\nTop scorer:", top["name"], "-", top["marks"])
```
