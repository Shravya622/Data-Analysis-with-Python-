## **Advanced Dictionary Practice**

**Problem 1** – **Count Occurrences**

**Given:**

words = ["apple", "banana", "apple", "orange", "banana", "apple"]

Create a dictionary that counts how many times each word appears.

**Expected structure:**

{'apple': 3, 'banana': 2, 'orange': 1}

```python
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
word_count={}
for word in words:
  if word in word_count:
    word_count[word]+=1
  else:
    word_count[word]=1
print(word_count)

```

**Problem 2** – **Filter Data**

**Given:**

students = {
    "Shravya": 90,
    "Anu": 75,
    "Rahul": 85,
    "Riya": 60
}

Create a new dictionary containing only students with marks ≥ 80.

```python
students = {
    "Shravya": 90,
    "Anu": 75,
    "Rahul": 85,
    "Riya": 60
}

filtered_students = {}

for name, marks in students.items():
    if marks >= 80:
        filtered_students[name] = marks

print(filtered_students)
```

**Problem 4** – **Nested Dictionary Practice**

Shravya scored 90

**Problem 5** – **Transform Dictionary**

Given:

prices = {
    "apple": 100,
    "banana": 40,
    "mango": 120
}

Create a new dictionary with 10% discount applied.

```python
prices = { "apple": 100, "banana": 40, "mango": 120 }
discount_price={}
for fruit,price in prices.items():
  discount_price[fruit]=price*0.9
print(discount_price)
```

**Mini Project** – **Inventory System**

Create a dictionary:

inventory = {
    "laptop": 5,
    "mouse": 20,
    "keyboard": 10
}

Tasks:

Print all items and quantities

Add a new item "monitor": 7

Reduce "mouse" quantity by 5

Print updated inventory

```python
inventory={
    "laptop":5,
    "mouse":20,
    "keyboard":10
}
for item,quantity in inventory.items():
  print(item,":",quantity)
inventory["monitor"]=7
inventory["mouse"]-=5

print("\nUpdated Inventory:")
for item,quantity in inventory.items():
    print(item,":",quantity)
```
